```python
df = pd.DataFrame(
    {
        'c1': [1,1,2,2],
        'c2': [5,5,7,8],
        'c3': [9, 10, 11, 12],
        'c4': [13, 14, 15, 16]
    }
)
df
```

```python
new_df = pd.pivot_table(df, values='c4', index=['c1', 'c2'], columns=['c3'], aggfunc="sum").reset_index()
new_df
```

```python
dd = pd.DataFrame(new_df.stack()).rename(columns={0: 'index'}).reset_index()
dd
```
